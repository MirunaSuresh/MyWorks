```python
for b in soup.find_all(href=True):
    attributes = b.attrs
    #print(attributes)
    if 'class' in attributes:
        if attributes['class']==['result-card__full-card-link']:
            print(attributes['href'])
            requestinternal = requests.get(attributes['href'])
            soupinternal = BeautifulSoup(requestinternal.content, 'html.parser')
            for div in soupinternal.find_all('section',class_="description"):
                for p in div.find_all('p'):
                    print(p.text)
                    
            
```
