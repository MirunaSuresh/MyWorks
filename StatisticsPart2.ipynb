```python
X = [0, 5, 10, 15, 20]
Y = [0, 7, 10, 13, 20]

# solve for a and b
def best_fit(X, Y):
    '''
    Return m, b
    '''
    x=np.array(X)
    y=np.array(Y)
    x2 = x**2
    xy =  x*y
    print(x2)
    print(xy)
    sumx= np.sum(x)
    print(sumx)
    sumy= np.sum(y)
    print(sumy)
    sumx2 = np.sum(x2)
    print(sumx2)
    sumxy = np.sum(xy)
    print(sumxy)
    m= (sumxy-(sumx*sumy))/(len(x)*( sumx2 - (sumx**2)))
    b= (sumy-(m*sumx))/len(x)
    
    
    return m, b

# solution
m, b = best_fit(X, Y)
print(m)
print(b)
print(np.polyfit(X,Y,1))
#best fit line:
#y = 0.80 + 0.92x

# plot points and fit line
_=plt.scatter(X,Y)
x= range(10)
y=[]
for i in range(10):
    y.append(m*x[i]+b)
_=plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import scipy.stats


def mean_confidence_interval(data, confidence=0.95):
    # Calculate m, h
    data = np.array(data)
    m, se = np.mean(data),scipy.stats.sem(data)
    h=stats.t.cdf(abs(confidence - m)/se,len(data)-1)
    
    return m, m-h, m+h

data = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
print(mean_confidence_interval(data))
```

```python
from scipy import stats
import matplotlib.pyplot as plt

X_main = stats.loggamma.rvs(5, size=500) + 5
```

```python
x,y=stats.probplot(X_main)[0]
plt.plot(x,y)
plt.title("Probplot for loggamma dist ")
plt.show()
```

```python
stats.probplot(X_main)[1]
```

```python
x,y = stats.probplot(X_main, dist=stats.norm)[0]
plt.plot(x,y)
plt.title('Probplot against normal distribution')

plt.show()
```

```python
xt,_=stats.boxcox(X_main)
x,y = stats.probplot(xt, dist=stats.norm)[0]
plt.plot(x,y)
plt.title('Boxcox')

plt.show()

```
