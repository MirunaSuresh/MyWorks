```python
import sqlite3
```

```python
sqlite_db = 'test_db.sqlite'
conn = sqlite3.connect(sqlite_db) 
c = conn.cursor()
```

```python

try:
    c.execute('DROP TABLE houses;')
except:
    print("no such table")
conn.commit()
```

```python
c.execute('CREATE TABLE if not exists houses (field1 INTEGER PRIMARY KEY, sqft INTEGER, bdrms INTEGER, age INTEGER, price INTEGER);')
conn.commit()
```

```python
last_sale = (None, 4000, 5, 22, 619000)
sql="insert into houses(field1,sqft,bdrms,age,price) values(?,?,?,?,?)"
c.execute(sql,last_sale)

conn.commit()
```

```python
c.execute("select * from houses")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
import pandas as pd
import sqlite3
from pandas.io import sql
```

```python
orders = pd.read_csv('../data/P12-ListOfOrders.csv', encoding = 'utf-8')
orders_break_down =  pd.read_csv('../data/P12-OrderBreakdown.csv', encoding = 'utf-8')
```

```python
orders.head()
```

```python
orders_break_down.head()
```

```python
orders.columns = [o.replace(' ', '_') for o in orders.columns.str.lower()]
```

```python
orders.head()
```

```python
orders_break_down.columns = [o.replace(' ', '_') for o in orders_break_down.columns.str.lower()]
```

```python
orders.dtypes
```

```python
orders_break_down.dtypes
```

```python
db_connection = sqlite3.connect('../DATA/eshop.db.sqlite')
orders.to_sql('orders', con = db_connection, if_exists = 'replace', index = False)
orders_break_down.to_sql('orders_break_down', con = db_connection, if_exists = 'replace', index = False)
```

```python
c = db_connection.cursor()
c.execute("select customer_name,count(*) as number_of_orders  FROM orders group by customer_name")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
#Select Number of Customers for Each Country
c.execute("select country,count(*) as number_of_orders_per_country  FROM orders group by country")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
#Select Number of Customers for Each Country
c.execute("select country,count(distinct customer_name) as number_of_orders_per_country FROM orders group by country")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
#  Select discount, sales, quantity for Each Order from orders_break_down Table
c.execute("select discount, sales, quantity  FROM orders_break_down ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select discount, sales, quantity, total price for Each Order from orders_break_down Table
c.execute("select discount, sales, quantity , sales*quantity-discount as total_price FROM orders_break_down ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select All Orders from orders_break_down Table Where Total Price Greater Than 100
c.execute("select discount, sales, quantity , sales*quantity-discount as total_price FROM orders_break_down where total_price>100 ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select All Customer And The Product They Have Bought
c.execute("select customer_name ,product_name FROM orders inner join orders_break_down  where orders.order_id = orders_break_down.order_id group by customer_name ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select Number of 'Furniture' Orders For Each Country
c.execute("select country ,count(product_name) FROM orders inner join orders_break_down  on orders.order_id = orders_break_down.order_id where category='Furniture' group by country ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select Number of 'Furniture' Orders For The Country Denmark
c.execute("select country ,count(product_name) FROM orders inner join orders_break_down  on orders.order_id = orders_break_down.order_id where category='Furniture' and country='Denmark' group by country ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select Total Sales With Discount and Without Discount for Each Country
c.execute("select country ,sum(sales*quantity-discount) as sales_with_discount, sum(sales*quantity) as sales_without_discount FROM orders inner join orders_break_down  on orders.order_id = orders_break_down.order_id  group by country ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select Total Quantity, Total Sales for Each Country
c.execute("select country ,sum(quantity), sum(sales) FROM orders inner join orders_break_down  on orders.order_id = orders_break_down.order_id  group by country ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select Top 3 Country Based on Sales
c.execute("select country ,sum(quantity), sum(sales) FROM orders inner join orders_break_down  on orders.order_id = orders_break_down.order_id  group by country order by sum(sales) desc limit 3")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select Bottom 3 Country Based On Quantities
c.execute("select country ,sum(quantity), sum(sales) FROM orders inner join orders_break_down  on orders.order_id = orders_break_down.order_id  group by country order by sum(quantity)  limit 3")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select Average Sales By Categroy For The Country 'France'
c.execute("select country ,category,avg(sales) FROM orders inner join orders_break_down  on orders.order_id = orders_break_down.order_id  where country='France' group by category ")
rows = c.fetchall()
for r in rows:
    print(r)
```

```python
# Select Country, Category and Total Sales Where Average Total Sales is The Highest
c.execute("select country ,category,sum(sales) FROM orders inner join orders_break_down  on orders.order_id = orders_break_down.order_id  group by category  order by avg(sales) desc limit 1 ")
rows = c.fetchall()
for r in rows:
    print(r)
```
